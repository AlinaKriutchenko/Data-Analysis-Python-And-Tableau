### Data for the Tableau map

```python
df2 = data
df2.insert(1, 'Split', '_')
df2["Cities"] = df2["AustralianPort"] + df2["Split"] + df2["ForeignPort"]
df2 = df2.groupby('Cities').sum()
df2 = df2.reset_index()
df2[['Aus','Foreign']] = df2.Cities.str.split("_",expand=True,)
df2 = df2.drop(['Cities', 'Year'], axis = 1)
#df2
```

### Find latitude and longitude for origin and destination city

```python
!pip install geopy 
!pip install Nominatim
from geopy.geocoders import Nominatim


aus = df2['Aus']
foreign = df2['Foreign']

lat1 = []
lon1 = []
lat2 = []
lon2 = []

# Get Latitude and Longitude
for i in range(0,353):
    geolocator = Nominatim(user_agent="Your_Name")
    
    location = geolocator.geocode(aus[i])
    lat1.append(location.latitude)
    lon1.append(location.longitude)
    
    location = geolocator.geocode(foreign[i])
    lat2.append(location.latitude)
    lon2.append(location.longitude)


    
# Combine lists    
list_for = pd.DataFrame(
    {'lat1': lat1,
     'lon1': lon1,
     'lat2': lat2,
     'lon2': lon2
    })
list_for


# Print Dataframe
df3 = pd.concat([df2, list_for], axis=1)
df3
```

### ML experiment

##### The test accuracy is extremely low as expected. ML is not suitable.

```python
d = data[['Month', 'AustralianPort', 'ForeignPort', 'Country', 'Year', 'TotalPax', 'TotalMail']]
d = pd.get_dummies(d)
d['Year'] = d['Year'].astype(np.int64)
d['TotalPax'] = d['TotalPax'].astype(np.int64)

# Load libraries
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn import metrics

# Variables
X = d.drop(['TotalPax'], axis = 1)
y = d['TotalPax']

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

# Create Decision Tree classifer 
clf = DecisionTreeClassifier()
# Fit Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
